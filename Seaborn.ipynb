## Seaborn Tutorial

TIPS Dataset

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

##### What could be the Tip based on other features?
* Tip is a dependent variable

#### Correlation with Heatmap

```python
df.corr() 

# Only the numeric columns are considered as part of correlation
# The diagonal values will be 1 as it is 1-to-1 relation
# Higher the correlation value, higher are the 2 feature related
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')

# 'reg' draws a probability density function PDF
# it also draws the Best-Fit line or Regression line
```

```python
# More than 2 variables are to be included

sns.pairplot(df)
```

```python
sns.pairplot(hue='sex', data=df)
```

```python
sns.pairplot(hue='smoker', data=df)
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False) 

# kde - kernel density estimation
# by default it is True
# it gives a relative percentage
# when specified as False, it gives the absolute value
```

### Categorical Plots

* boxplot
* violinplot
* countplot
* bar plot

```python
sns.countplot('sex',data=df) # we specify only one axis, the other axis will be count
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='time',data=df)
```

```python
# For bar plot we define both the X-Axis and Y-Axis

sns.barplot(x='smoker', y='total_bill', data=df)
```

#### Box Plot

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
# hue to impose another feature on top of a boxplot

sns.boxplot(x='total_bill',y='day',data=df, palette='rainbow', hue='smoker')
```

#### Violin Plot

```python
sns.violinplot(x='total_bill', y='day', data=df)
```
